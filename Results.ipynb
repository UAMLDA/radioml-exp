# Results from Adversarial RadioML

This notebook summarizes the results from adversarial attacks against the RadioML dataset. The result logger class in `arml/performances.py` has the performances for the accuracy, AUC and perplexity.

```python
import pickle 
import matplotlib.pylab as plt
plt.style.use('bmh')
```

# FastGradientMethod Epsilon Variation 

```python
data = pickle.load(open('outputs/aml_fgsm_vtcnn2_vtcnn2_scenario_A_radioml-reverse_sigmoid.pkl', 'rb'))
logger = data['result_logger']

plt.figure()
for i, eps in enumerate(logger.epsilons): 
    plt.plot(logger.snrs, logger.accuracy[:, i]/logger.count, marker='o', label=''.join(['Epsilon=', str(eps)]))
plt.legend()
plt.ylabel('Accuracy')
plt.savefig('outputs/radioml_fgsm_accuracy_epsilons-reverse_sigmoid.pdf')

plt.figure()
for i, eps in enumerate(logger.epsilons): 
    plt.plot(logger.snrs, logger.aucs[:, i]/logger.count, marker='o', label=''.join(['Epsilon=', str(eps)]))
plt.legend()
plt.ylabel('AUC')
plt.savefig('outputs/radioml_fgsm_auc_epsilons-reverse_sigmoid.pdf')
```

# Multiple Attacks  

```python
data = pickle.load(open('outputs/aml_vtcnn2_vtcnn2_scenario_A_radioml.pkl', 'rb'))
logger = data['result_logger']

plt.figure()
plt.plot(logger.snrs, logger.accuracy, label='Benign', marker='o')
plt.plot(logger.snrs, logger.accuracy_fgsm, label='FGSM', marker='o')
plt.plot(logger.snrs, logger.accuracy_deep, label='DeepFool', marker='o')
plt.plot(logger.snrs, logger.accuracy_pgd, label='PGD', marker='o')
plt.legend()
plt.xlabel('dB')
plt.ylabel('Accuracy')
plt.savefig('outputs/radioml_adversarial_accuracy.pdf')

plt.figure()
plt.plot(logger.snrs, logger.perplexity, label='Benign', marker='o')
plt.plot(logger.snrs, logger.perplexity_fgsm, label='FGSM', marker='o')
plt.plot(logger.snrs, logger.perplexity_deep, label='DeepFool', marker='o')
plt.plot(logger.snrs, logger.perplexity_pgd, label='PGD', marker='o')
plt.legend()
plt.xlabel('dB')
plt.ylabel('Perplexity')
plt.savefig('outputs/radioml_adversarial_perplexity.pdf')


plt.figure()
plt.plot(logger.snrs, logger.aucs, label='Benign', marker='o')
plt.plot(logger.snrs, logger.aucs_fgsm, label='FGSM', marker='o')
plt.plot(logger.snrs, logger.aucs_deep, label='DeepFool', marker='o')
plt.plot(logger.snrs, logger.aucs_pgd, label='PGD', marker='o')
plt.legend()
plt.xlabel('dB')
plt.ylabel('AUC')
plt.savefig('outputs/radioml_adversarial_auc.pdf')
```
