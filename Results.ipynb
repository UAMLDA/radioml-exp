# Results from Adversarial RadioML

This notebook summarizes the results from adversarial attacks against the RadioML dataset. The result logger class in `arml/performances.py` has the performances for the accuracy, AUC and perplexity. 

```python
import pickle 
import matplotlib.pylab as plt
plt.style.use('seaborn-colorblind')
```

```python
data = pickle.load(open('outputs/aml_vtcnn2_vtcnn2_scenario_A_radioml.pkl', 'rb'))
logger = data['result_logger']
```

```python
plt.figure()
plt.plot(logger.accuracy, label='Benign', marker='o')
plt.plot(logger.accuracy_fgsm, label='FGSM', marker='o')
plt.plot(logger.accuracy_deep, label='DeepFool', marker='o')
plt.plot(logger.accuracy_pgd, label='PGD', marker='o')
plt.legend()
plt.xlabel('dB')
plt.ylabel('Accuracy')
plt.savefig('outputs/radioml_adversarial_accuracy.pdf')

plt.figure()
plt.plot(logger.perplexity, label='Benign', marker='o')
plt.plot(logger.perplexity_fgsm, label='FGSM', marker='o')
plt.plot(logger.perplexity_deep, label='DeepFool', marker='o')
plt.plot(logger.perplexity_pgd, label='PGD', marker='o')
plt.legend()
plt.xlabel('dB')
plt.ylabel('Perplexity')
plt.savefig('outputs/radioml_adversarial_perplexity.pdf')


plt.figure()
plt.plot(logger.aucs, label='Benign', marker='o')
plt.plot(logger.aucs_fgsm, label='FGSM', marker='o')
plt.plot(logger.aucs_deep, label='DeepFool', marker='o')
plt.plot(logger.aucs_pgd, label='PGD', marker='o')
plt.legend()
plt.xlabel('dB')
plt.ylabel('AUC')
plt.savefig('outputs/radioml_adversarial_auc.pdf')
```
